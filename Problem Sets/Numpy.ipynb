12. 
- Create a 2D array and print its shape as well as ndim. (Any shape is fine)

- Create a 3D array and print its shape as well as ndim.

- Create a 4D array and print its shape as well as ndim. *It's hard to visualize 4-Dimensions but numpy can do it with ease!*

```python
weight_g = [10.5, 20.3, 12.3, 16.8, 20.3]
svl_cm = [5.2, 30.5, 6.3, 8.1, 28.7]
species = ['d_punctatus', 'c_coluber', 's_dekayi', 'c_coluber', 't_sirtalus']
tail_cm = [2.1, 7.3, 2.5, 3.3, 6.5]

np_weight_g = np.array(weight_g)
np_svl_cm = np.array(svl_cm)
np_species = np.array(species)
np_tail_cm = np.array(tail_cm)

#1 dimensional
print(np_weight_g.shape)
print(np_weight_g.ndim)
print()

#standard 2d array

# _2d_array = np.array([[0, 2, 4, 4], [7, 4, 3, 5]])
# print (_2d_array.shape)
# print(_2d_array.ndim)
# print()

#2-dimensional, column_stack changes the 2 rows of 1-dimensional arrays into 2 columns of 1 dimensional arrays
np_species_weight = np.array([[np_species],[np_weight_g]])
print(np_species_weight)
print()
new_np_species_weight = np.column_stack(np_species_weight)
print(new_np_species_weight.shape)
print(new_np_species_weight.ndim)

#3-dimensional
np_species_svl_weight = np.array([[np_species], [np_svl_cm],[np_weight_g]])
print(np_species_svl_weight)
print()
print(np_species_svl_weight.shape)
print(np_species_weight.ndim)

#4-dimensional
np_all = np.array([[[np_species], [np_svl_cm], [np_tail_cm], [np_weight_g]]])
print(np_all)
print(np_all.shape)
print(np_all.ndim)
```
